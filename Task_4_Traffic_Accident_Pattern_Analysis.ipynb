```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)

data = {
    "Weather": np.random.choice(["Clear", "Rain", "Fog", "Snow"], 500),
    "Road_Condition": np.random.choice(["Dry", "Wet", "Icy"], 500),
    "Time_of_Day": np.random.choice(["Morning", "Afternoon", "Evening", "Night"], 500),
    "Severity": np.random.choice(["Minor", "Major", "Fatal"], 500),
    "Location": np.random.choice(["Highway", "City", "Rural"], 500)
}

df = pd.DataFrame(data)
df.head()
```

```python
sns.countplot(x="Weather", data=df)
plt.title("Accidents by Weather")
plt.show()
```

```python
sns.countplot(x="Road_Condition", data=df)
plt.title("Accidents by Road Condition")
plt.show()
```

```python
sns.countplot(x="Time_of_Day", data=df)
plt.title("Accidents by Time of Day")
plt.show()
```

```python
pivot = pd.crosstab(df["Location"], df["Severity"])

sns.heatmap(pivot, annot=True, cmap="Reds")
plt.title("Accident Hotspots Heatmap")
plt.show()
```
